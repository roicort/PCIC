```python
import pandas as pd
import os 
import plotly.express as px

# Load the data (get only csv files)

data_dir = "itslive-data"
files = [f for f in os.listdir(data_dir) if f.endswith('.csv')]
print(files)

for file in files:

    df = pd.read_csv(data_dir+"/"+file)
    df['mid_date'] = pd.to_datetime(df['mid_date'])

    print(df.head())

    fig = px.scatter(df, x='mid_date', y=' v [m/yr]', title=file, trendline='lowess')
    fig.show()
    
    
```
