## Análisis léxico

Utiliza un programa en Python para analizar las partes de la oración de cada una de las palabras del siguiente texto y su conteo.

Montecristo esconde un fantástico tesoro de perlas y diamantes en bruto que bien vale la pena visitar. Cerca, el viejo sabio observa cómo el joven corre ligero mientras el viento fresco despierta sueños olvidados. Al final del sendero, la luz tenue revela secretos que nadie jamás supo descifrar.

1. Etiqueta cada palabra con su categoría gramatical y su lema. Marca todas aquellas que consideres que tienen una etiqueta errónea y señala cuál debe ser la correcta.
2. Haz una lista de los tokens y lemas, con sus frecuencias.

```python
import spacy
from collections import Counter

# Cargar el modelo de spaCy para español
nlp = spacy.load("es_core_news_sm")

# Texto a analizar
texto = ("Montecristo esconde un fantástico tesoro de perlas y diamantes en bruto que bien vale la pena visitar. "
         "Cerca, el viejo sabio observa cómo el joven corre ligero mientras el viento fresco despierta sueños olvidados. "
         "Al final del sendero, la luz tenue revela secretos que nadie jamás supo descifrar.")

# Procesar el texto con spaCy
doc = nlp(texto)

# Etiquetar cada palabra con su categoría gramatical y lema
tokens = [(token.text, token.lemma_, token.pos_) for token in doc]

# Mostrar las etiquetas y lemas
for token in tokens:
    print(f"Palabra: {token[0]}, Lema: {token[1]}, Categoría gramatical: {token[2]}")
```

Palabra: viejo, Lema: viejo, Categoría gramatical: ADJ -> No se usa como adjetivo


```python
# Contar la frecuencia de los tokens y lemas
token_freq = Counter([token.text for token in doc])
lemma_freq = Counter([token.lemma_ for token in doc])

# Mostrar las frecuencias de los tokens y lemas
print("\nFrecuencia de tokens:")
for token, freq in token_freq.items():
    print(f"{token}: {freq}")

print("\nFrecuencia de lemas:")
for lemma, freq in lemma_freq.items():
    print(f"{lemma}: {freq}")
```
